```python
arr = np.array([4,2,6,8],dtype = 'int64')
print(arr*2,"\n")
print(arr*arr,"\n")
print(arr//2,"\n")
print(arr/2,"\n")
print(2/arr,"\n")
print(2//arr,"\n")
print(arr**2,"\n")
print(np.sqrt(arr),"\n")
print(np.exp(arr),"\n")
print(np.mean(arr),"\n")
print(np.sum(arr),"\n")
print(np.cumsum(arr),"\n")
print(np.sort(arr),"\n")

arr2 = np.array([0,1,3,9])

print(arr > arr2)
```

```python
namesArr = np.array(["Sarfraz","Zeeshan","Sarwar","Sarfraz","Aslam"])
numArr = np.random.randn(5,3)   # jitne array namesArr mai define kiya hai utne hi array(dimension) bnaege yhan manipulate krne k liye ..

print(numArr,"\n")
print(namesArr,"\n")
print(namesArr == "Sarfraz","\n")
print(numArr[ namesArr == "Sarfraz"],"\n")
print(numArr[ ~(namesArr == "Sarfraz")],"\n") #~(tidle) = not
print(np.unique(namesArr))   # sort and neglate multiple(repeated) element
```
